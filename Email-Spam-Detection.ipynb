__Replacing Null values__

```python
df = df.where((pd.notnull(df)), '')
```

```python
df
```

```python
df.head()
```

```python
df = df.rename(columns={'v1' : 'Category', 'v2' : 'Messages'})
```

```python
df.head()
```

__Label Encoding__

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder = LabelEncoder()
df['Category'] = label_encoder.fit_transform(df['Category'])
```

__Spltting Data__

```python
x = df['Messages']
y = df['Category']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=50, test_size=0.2)
```

__Feature Extraction__

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
features_extraction = TfidfVectorizer(min_df=1,stop_words='english',lowercase=True)
x_train_features = features_extraction.fit_transform(x_train)
x_test_features = features_extraction.transform(x_test)
```

```python
y_train = y_train.astype(int)
y_test = y_test.astype(int)
```

__Model Training__

```python
from sklearn.linear_model import LogisticRegression
```

```python
lo = LogisticRegression()
lo.fit(x_train_features,y_train)
```

__Evaluating__

```python
from sklearn.metrics import accuracy_score
```

```python
prediction = lo.predict(x_train_features)
accuracy = accuracy_score(y_train,prediction)
```

```python
accuracy
```

__Predicting New Value__

```python
mail = ["I've been searching for the right words to thank you for this breather. I promise i wont take help for granted and will fulfill my promise. "]
```

```python
features = features_extraction.transform(mail)
```

```python
predictions = lo.predict(features)
print(predictions)

if (predictions[0]==1):
    print('Not Spam')
else:
    print('Spam')
```
